```python
import matplotlib.pyplot as plt
```

```python
languages = 'Python', 'Java', 'PHP', 'Javascript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
colors = ['#FA8F0A','#F0DA32','#61F527','#18DE9D','#3594FF','#8B5A8C']
```

```python
plt.pie(popularity, labels=languages, colors=colors)
plt.title('Popularity of Programming Languages in 2019')
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%')
plt.show()
```

```python
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True)
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140)
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame= True)
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True)
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True, wedgeprops={'linewidth':3})
plt.show()
```

```python
plt.pie(popularity, labels=languages, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=140,
        frame=True, rotatelabels=True, wedgeprops = {'linewidth': 3}, radius=2)
plt.show()
```

## Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values = [82,76,24,40,67,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52]
```

```python
plt.hist(values, histtype='bar')
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='k')
```

```python
plt.hist(values, histtype='bar', edgecolor='k')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('frequency')
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='k', label='Test score data')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='k', label='Test score data', range=[0,100])
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

```python
plt.hist(values, histtype='bar', edgecolor='k', label='Test score data', range=[0,100], orientation='horizontal')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

```python
values1 = [42,36,24,40,67,62,45,58,41,32,48,49,38,57,42,72,57,56,56,43]

plt.hist(values, histtype='bar', edgecolor='k', label='Test1 score data',alpha=0.8, color='c')
plt.hist(values1, histtype='bar', edgecolor='k', label='Test2 score data',alpha=0.5, color='r')
plt.title('Histogram of Scores')
plt.xlabel('Scores')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

## BOX PLOTS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cancer.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
plt.boxplot(df.radius_mean)
```

```python
plt.boxplot(df.radius_mean, showmeans=True, vert=False, sym='b+')
```

```python
df.boxplot(column = 'area_mean', by = 'diagnosis');
plt.title('')
```

```python
sns.boxplot(x='diagnosis', y='area_mean', data=df)
```

```python
#Step9: Graph the boxplot using matplotlib
malignant = df[df['diagnosis']=='M']['area_mean']
benign = df[df['diagnosis']=='B']['area_mean']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot([malignant,benign], labels=['M', 'B'])
```
